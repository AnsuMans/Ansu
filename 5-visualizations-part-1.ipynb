```python
import numpy as np # linear algebra
import pandas as pd 
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
        import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder 
plt.style.use(style='seaborn')

paysim_data = pd.read_csv("../input/paysim1/PS_20174392719_1491204439457_log.csv",encoding="ISO-8859-1",error_bad_lines=False)
paysim_data.head(2)

x = paysim_data['type'].value_counts().index
y = paysim_data['type'].value_counts().values

f = plt.figure(1,figsize=(16,6))
ax1 = f.add_subplot(1,2,1)
ax1.title.set_text('Type')
_ =ax1.bar(x,y)

z = paysim_data['amount'].value_counts().index

ax2 = f.add_subplot(1,2,2)
ax2.title.set_text('Amount')
_ =ax2.boxplot(z)

corr = paysim_data.corr()
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)

```
